# PyPDF

**PyPDF is a pure-Python library for:**


* Reading PDFs
* Splitting & merging
* Extracting text & metadata
* Encrypting/decrypting
* Basic manipulation (pages, rotation, etc.)


It does NOT:

Render PDFs visually (like a viewer)

Reliably extract structured data (tables are messy)

*Core Concepts**
1. Reader vs Writer

PdfReader → opens and reads PDFs

PdfWriter → creates or modifies PDFs

```python
from pypdf import PdfReader, PdfWriter
```

```python
#Reading a PDF

reader = PdfReader("/kaggle/input/datasets/nallagantisharath/tax-data/1120.pdf")

print(len(reader.pages))  # number of pages

page = reader.pages[0]
text = page.extract_text()

print(text)
```

```python
# Splitting PDFs

reader = PdfReader("/kaggle/input/datasets/nallagantisharath/tax-data/i1065.pdf")
writer = PdfWriter()

writer.add_page(reader.pages[0])

with open("output.pdf", "wb") as f:
    writer.write(f)
```

```python
# Merging PDFs



from pypdf import PdfWriter

writer = PdfWriter()

for pdf in ["/kaggle/input/datasets/nallagantisharath/tax-data/1120.pdf", "/kaggle/input/datasets/nallagantisharath/tax-data/i1065.pdf"]:
    writer.append(pdf)

with open("merged.pdf", "wb") as f:
    writer.write(f)
```

```python
# Encrypt & Decrypt


# Encrypt:

writer = PdfWriter()
writer.append_pages_from_reader(reader)

writer.encrypt("mypassword")

with open("secure.pdf", "wb") as f:
    writer.write(f)



# Decrypt:
reader = PdfReader("secure.pdf")

if reader.is_encrypted:
    reader.decrypt("mypassword")
```

```python
# Extract Metadata

reader = PdfReader("file.pdf")

meta = reader.metadata
print(meta.title)
print(meta.author)
```
