# LangChain 

**LangChain as a framework that helps you connect LLMs with external data and tools**

**LangChain = Framework for building applications using Large Language Models (LLMs)**

It helps combine: LLM with documents+ databases+ APIs+ memory+ prompts as one system.


LangChain is an open-source framework (mostly in Python, with a JavaScript/TypeScript version too) that makes it much easier to build powerful applications powered by large language models (LLMs) like GPT-4, Claude, Gemini, Llama, Grok, and many others

**LLMs alone cannot:**

1.access documents

2.search databases

3.remember conversation

4.call APIs

5.execute tools

LangChain provides components to solve this.


**In simple terms**
Imagine you want to create an AI app that does more than just chat:

Answers questions using your company's private documents
Searches the internet in real time
Remembers previous conversations
Uses tools (calculator, code interpreter, APIs, databases)
Makes multi-step decisions (like an "AI agent" that plans and acts)

Doing all of this from scratch with just the OpenAI/Anthropic/etc. API is quite painful — you have to write a lot of glue code for prompts, memory, tool calling, error handling, chaining steps, etc.
LangChain gives you ready-made building blocks so you can snap these pieces together quickly instead of coding everything yourself.



**Core idea**: "Chains" + "Agents"

**Chains** → sequences of steps (e.g., prompt → LLM → parse output → another prompt → tool call → LLM again)
**Agents** → smarter systems that decide which tools to use and in what order, often using reasoning loops like ReAct ("think → act → observe → think…")

```python
!pip install langchain
!pip install -q langchain langchain-community langchain-text-splitters
!pip install -q sentence-transformers chromadb pypdf transformers

from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/kaggle/input/datasets/nallagantisharath/tax-data/1120.pdf")
docs = loader.load()


```

**Text Splitters**

LLMs cannot process very long documents.

Text splitters break documents into smaller chunks.


Example:

```python

from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_documents(docs)
```

**Embeddings**

Embeddings convert text into numerical vectors.

Example:

"efile Form 1120"

becomes

[0.12, -0.91, 0.33, 0.88 ...]

These vectors allow semantic search

Example:

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-base-en-v1.5"
)
```
