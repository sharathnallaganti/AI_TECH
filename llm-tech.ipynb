**Attention Mechanism** in LLMs It allows the model to dynamically focus on relevant parts of the input sequence when processing or generating each token.



#### Self-Attention (Scaled Dot-Product Attention)
This is the basic building block used in LLMs (decoder-only transformers).
For a sequence of tokens with embeddings X (shape: batch × sequence_length × d_model):



The model creates three learned projections:

**Query (Q):** "What am I looking for?"

**Key (K):** "What do I have to offer?"

**Value (V):** "What information should I pass on?"


Mathematically: $$Q = X W_Q, \quad K = X W_K, \quad V = X W_V$$
Then, Scaled Dot-Product Attention:
$$\text{Attention}(Q, K, V) = \text{softmax}\left( \frac{Q K^T}{\sqrt{d_k}} \right) V$$



$  Q K^T  $: Computes similarity (dot product) between every pair of tokens.

**Scaling** by $  \sqrt{d_k}  $: Prevents the dot products from becoming too large (which would make softmax too peaked and gradients vanish).

**Softmax:** Turns similarities into probabilities (attention weights) that sum to 1.

**Multiply by V**: Weighted sum → context-aware representation for each token.

**What is ChromaDB?**

ChromaDB is a vector database designed for AI applications.

It stores embeddings (vectors) and allows semantic search.

Instead of searching by keywords, it searches by meaning.

**Example:**

Query:

How do I efile Form 1120?

ChromaDB can find documents that mention:

IRS Modernized eFile system
corporate tax return submission

even if the exact words don't match.

**top 5 alternatives to ChromaDB:**

**Qdrant** – best balance of simplicity, speed, and scalability

**Pinecone** – fully managed, production-ready vector database

**Milvus** – highly scalable, distributed system for massive datasets

**Weaviate** – feature-rich with built-in AI and hybrid search

**FAISS** – ultra-fast local vector search library (not a full DB)

**Why Vector Databases Exist**

Normal databases search like this:

SQL search → exact match

Example:

SELECT * WHERE text LIKE "efile"

But AI applications need semantic similarity.

Vector databases allow:

Meaning-based search

Example:

"efile 1120"
≈
"submit corporate tax return electronically"

**Quantization**
for Large Language Models (LLMs) reduces the precision of model weights and activations—typically from 16-bit float (FP16) to 8-bit (INT8) or 4-bit (INT4) integers. This optimization drastically shrinks model size and speeds up inference, enabling large models to run on consumer hardware (GPUs, laptops, phones) with minimal accuracy loss.
