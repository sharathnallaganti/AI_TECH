# Model Context Protocol

MCP is an open protocol that standardizes how AI applications (hosts) discover and interact with external services, data sources, and tools. 

It's often described as a "USB-C for AI" because it decouples AI models from the systems they use, enabling plug-and-play interoperability.



**Core Problem it Solves:**

Before MCP, every AI app needed custom integrations for every tool (Google Drive, GitHub, Slack, databases, etc.). This was messy, insecure, and non-portable. MCP creates one standard interface.






#  Core Architectural Components

### MCP Host:
   The user-facing application (IDE, desktop app, agent framework, web UI) that orchestrates AI workflows.

### MCP Client:
   Embedded inside the host. Manages connections,sends JSON-RPC requests, and handles responses from servers.

### MCP Server:

   A lightweight, independent process that exposes Exposes data/tools Runs locally or remotely.

# Protocol Primitives

### MCP exposes three core building blocks:

**Resources:**
Read-only data endpoints (files, DB rows, API responses). Supports URI-based access and live subscriptions.

**Tools:** 
Callable functions with JSON Schema definitions. Enables safe, structured function calling by the LLM.

**Prompts:** Reusable, parameterized prompt templates that servers can share with clients.

**Sampling (optional):** Servers can ask the client's LLM to generate text, enabling bidirectional AI workflows.

**Roots:** Explicitly define filesystem/data boundaries for security and scoping.
