**Transformers** are a deep learning architecture, based on the self-attention mechanism, designed to process sequential data (like text, audio, or time-series) in parallel. 
They enable models to weigh the importance of different parts of an input simultaneously, making them highly effective for tasks like language understanding, generation, and beyond.

**Transformers** aren’t just “upgraded” LSTMs or GRUs  They’re a fundamentally different architecture that solves the same problem in a completely new way.


### LSTM / GRU vs. Transformer


| Feature | LSTM / GRU | Transformer |
| :--- | :--- | :--- |
| **Data Reading** | Reads data sequentially (word → word → word) | Looks at the entire sequence at once |
| **Context Mechanism** | Carries context forward in a hidden state | Uses attention to instantly link related parts, no matter how far apart |
| **Training Speed** | Can't easily parallelize training | Fully parallelizable → much faster to train on GPUs |
| **Sequence Length** | Struggles with very long sequences (context "fades") | Handles long-range dependencies naturally |


## Why Transformers Exist

To understand transformers:

You first need to understand: **What problem were they solving?****

Before transformers, NLP used:

 * RNNs
 * LSTMs
 * GRUs

These processed text sequentially.

Example sentence:

“The cat sat on the mat.”

RNN flow: The → cat → sat → on → the → mat

Token by token.

**This caused 3 big problems**

**Problem 1 — Poor Long-Range Dependency Modeling**

**Example:**

“The animal that chased the cat near the building was hungry.”

Suppose model predicts:

   “was”

**To understand grammar:**

the model must remember:

    “animal”

far back in the sequence.

RNNs struggle because information gets diluted.

**This creates: Vanishing Gradient Problem**

Earlier information fades.

Think: token1 → token2 → token3 → ... → token100



**Transformer solution:**

Look anywhere directly.

No sequential bottleneck.


**Problem 2 — Sequential Computation**

RNN: h1 → h2 → h3 → h4

Must compute in order.

Cannot parallelize efficiently.

GPU utilization suffers.

Training slow.

**Transformer:**

All tokens processed simultaneously.

Example: [The, cat, sat, on, mat] processed together , Huge speedup and This is one reason scaling became possible.


**Problem 3 — Fixed Memory Bottleneck**

RNN compresses sentence meaning into hidden state.

Like: sentence → compressed memory vector hence Information bottleneck.


**Transformers instead:**

dynamically retrieve information using attention hence No fixed bottleneck.

**Attention:** A mechanism that helps the **model focus on the most important words in a sentence** instead of treating every word equally.
It decides which words are most relevant for understanding the current word. This helps the model understand meaning and relationships in text.



**Self Attention:** A type of attention where **each word looks at all other words in the same sentence** to understand context better. 
It helps words understand their meaning based on surrounding words. For example, it helps the model know which earlier word a pronoun refers to.



**Multi Head Attention:** A method where the **model uses multiple attention processes at the same time to learn different relationships in text**. One part may focus on grammar while another focuses on meaning or context. This helps the model understand language more deeply and accurately.

**Masked Multi Head Attention:** A type of multi-head attention used in language models where the model can only look at current and previous words, not future words. 
This prevents cheating during text generation and helps the model predict the next word step by step. Multiple attention heads work together while following this restriction to learn different patterns in the sequence.
