**Transformers** are a deep learning architecture, based on the self-attention mechanism, designed to process sequential data (like text, audio, or time-series) in parallel. 
They enable models to weigh the importance of different parts of an input simultaneously, making them highly effective for tasks like language understanding, generation, and beyond.

**Transformers** aren’t just “upgraded” LSTMs or GRUs  They’re a fundamentally different architecture that solves the same problem in a completely new way.


### LSTM / GRU vs. Transformer


| Feature | LSTM / GRU | Transformer |
| :--- | :--- | :--- |
| **Data Reading** | Reads data sequentially (word → word → word) | Looks at the entire sequence at once |
| **Context Mechanism** | Carries context forward in a hidden state | Uses attention to instantly link related parts, no matter how far apart |
| **Training Speed** | Can't easily parallelize training | Fully parallelizable → much faster to train on GPUs |
| **Sequence Length** | Struggles with very long sequences (context "fades") | Handles long-range dependencies naturally |


## Why Transformers Exist

To understand transformers:

You first need to understand: **What problem were they solving?****

Before transformers, NLP used:

 * RNNs
 * LSTMs
 * GRUs

These processed text sequentially.

Example sentence:

“The cat sat on the mat.”

RNN flow: The → cat → sat → on → the → mat

Token by token.

**This caused 3 big problems**

**Problem 1 — Poor Long-Range Dependency Modeling**

**Example:**

“The animal that chased the cat near the building was hungry.”

Suppose model predicts:

   “was”

**To understand grammar:**

the model must remember:

    “animal”

far back in the sequence.

RNNs struggle because information gets diluted.

**This creates: Vanishing Gradient Problem**

Earlier information fades.

Think: token1 → token2 → token3 → ... → token100



**Transformer solution:**

Look anywhere directly.

No sequential bottleneck.


**Problem 2 — Sequential Computation**

RNN: h1 → h2 → h3 → h4

Must compute in order.

Cannot parallelize efficiently.

GPU utilization suffers.

Training slow.

**Transformer:**

All tokens processed simultaneously.

Example: [The, cat, sat, on, mat] processed together , Huge speedup and This is one reason scaling became possible.


**Problem 3 — Fixed Memory Bottleneck**

RNN compresses sentence meaning into hidden state.

Like: sentence → compressed memory vector hence Information bottleneck.


**Transformers instead:**

dynamically retrieve information using attention hence No fixed bottleneck.

**Attention:** A mechanism that helps the **model focus on the most important words in a sentence** instead of treating every word equally.
It decides which words are most relevant for understanding the current word. This helps the model understand meaning and relationships in text.



**Self Attention:** A type of attention where **each word looks at all other words in the same sentence** to understand context better. 
It helps words understand their meaning based on surrounding words. For example, it helps the model know which earlier word a pronoun refers to.



**Multi Head Attention:** A method where the **model uses multiple attention processes at the same time to learn different relationships in text**. One part may focus on grammar while another focuses on meaning or context. This helps the model understand language more deeply and accurately.

**Masked Multi Head Attention:** A type of multi-head attention used in language models where the model can only look at current and previous words, not future words. 
This prevents cheating during text generation and helps the model predict the next word step by step. Multiple attention heads work together while following this restriction to learn different patterns in the sequence.


## Transformer Architecture





## Core Transformer Pipeline (High Level)

**tokens** --> **embeddings** --> **positional encoding** --> **self-attention**--> **feed-forward network** --> **repeat many layers** --> **next-token prediction**



**Tokenization**
It splits the sentence into tokens such as: The, cat, chased, the, small, mouse.

**Embeddings**

Each token is converted into a numerical vector. The model cannot understand words directly, so “cat” becomes a vector of numbers.

**Positional Encoding**

Since the Transformer reads all tokens at once, it needs position information. This tells the model that “cat” is second, “chased” is third, and “mouse” is sixth.

**Multi-Head Self-Attention**

Each word looks at other words to understand context. For example, “chased” connects strongly with “cat” and “mouse.” “small” connects strongly with “mouse.” Multiple heads learn different relationships at the same time.

**Add and Layer Normalization**

The model keeps the original information and stabilizes the values so training remains smooth.

**Feed Forward Network**

After attention finds relationships, the feed-forward network processes each token’s information further. You can think of attention as gathering context and feed-forward as refining or thinking over that context.

**Repeated Transformer Blocks**
The attention and feed-forward steps are repeated many times. Early layers may learn simple relationships, while deeper layers learn more abstract meaning and reasoning patterns.

**Output Layer**
Finally, the model produces an output. In language models, it usually predicts the next token.




Sentence:

> “The cat chased the small mouse”

Think of the Transformer as:

> Read sentence → understand relationships → refine meaning → predict output

### 1. Tokenization

The model first breaks the sentence into tokens.

Input:

> “The cat chased the small mouse”

Tokens:

> [The] [cat] [chased] [the] [small] [mouse]

Why?

Because the model cannot directly process text.

Example:
Instead of words, the computer sees IDs:

> the → 101,  cat → 205, chased → 876, the → 101, small → 432,mouse → 901

Now everything becomes numbers.

### 2. Embeddings

Each token ID becomes a vector of numbers representing meaning.

Example: word --> tocken --> embedding

“The”   --> 101 → [0.2, 0.8, 0.1]
“cat”   --> 205 → [0.9, 0.3, 0.7]
“mouse” --> 901 → [0.4, 0.6, 0.8]

You do not need to memorize numbers.

Think:

> embedding = numerical meaning of a word

Example intuition:

“cat” and “dog” will have similar embeddings because they are semantically related.

### 3. Positional Encoding

Problem:

Transformers process everything simultaneously.

Without position:

> “The cat chased mouse”

and

> “mouse chased cat The”

would look similar.

So position is added.

Example:

“The” at position 1
“cat” at position 2
“chased” at position 3

Now model knows order.

Meaning becomes:

> “cat” + “I am word number 2”

### 4. Self Attention

Now comes the important part.

Suppose we focus on:

> “chased”

The model asks:

> Which words are important for understanding “chased”?

It looks at:

The → small importance
cat → high importance
chased → medium importance
the → low importance
small → low importance
mouse → high importance

Why?

Because:

> cat chased mouse

contains the main meaning.

So “chased” pays more attention to:

> cat and mouse

Another example:

Word:

> “small”

Attention:

The → low
cat → low
mouse → high

because:

> small mouse

makes sense.

This helps the model understand relationships.

### 5. Multi Head Attention

Instead of one attention process:

The Transformer runs many attention heads at once.

Imagine 3 heads.

Head 1 focuses on grammar:

> cat → chased

Head 2 focuses on adjective relationships:

> small → mouse

Head 3 focuses on sentence meaning:

> cat chased mouse

So multiple perspectives are learned simultaneously.

Think:

> several experts analyzing the same sentence differently

### 6. Masked Multi Head Attention

Used during text generation.

Suppose model predicts:

> “small”

Input so far:

> “The cat chased the”

The model can see:

The
cat
chased
the

But cannot see:

small
mouse

Why?

Because during generation:

> future words must stay hidden

Otherwise model would cheat.

So prediction happens step by step.

Example:

“The cat chased the ___”

Model predicts:

> small

Then:

“The cat chased the small ___”

Model predicts:

> mouse

### 7. Feed Forward Network

Attention finds relationships.

Now model processes information more deeply.

Example:

After attention:

> cat ↔ chased ↔ mouse

Feed-forward refines understanding:

> cat is actor
> mouse is target
> chased is action

Think:

Attention:

> gather information

Feed Forward:

> think about information

### 8. Residual Connections

Problem:

Deep networks lose information.

Solution:

Keep original information.

Instead of:

> only new output

model does:

> old information + new information

Example:

Original understanding of “cat” is preserved while adding new context.

This improves stability.

### 9. Layer Normalization

Numbers inside network may become unstable.

Layer normalization keeps values balanced.

Think:

> keeps learning stable and prevents chaos

### 10. Repeat Transformer Blocks

This whole process repeats many times.

Layer 1 learns:

> small mouse

Layer 5 learns:

> cat chased mouse

Layer 20 learns:

> overall sentence meaning

Each layer refines understanding.

### 11. Output Prediction

Now model predicts next word.

Input:

> “The cat chased the small mouse because”

Model estimates probabilities:

> it = 45%
> the = 10%
> was = 25%

Chooses likely next token.
