# Normalization

## What is Normalization?

Normalization is the process of transforming data into a standard scale so that different features or values have comparable magnitudes.

The main goal is:

- Reduce scale differences
- Stabilize training
- Improve convergence
- Prevent exploding or vanishing values

Example:

Without normalization:

$$
[2,\ 20,\ 200,\ 2000]
$$

The values vary significantly.

After normalization:

$$
[0.001,\ 0.01,\ 0.1,\ 1]
$$

Now the values are on a comparable scale.

# Why Do We Need Normalization?

Consider two features:

$$
Age = [20,25,30]
$$

$$
Salary = [50000,70000,100000]
$$

Salary values dominate because they are much larger.

A neural network may focus excessively on Salary and ignore Age.

Normalization ensures all features contribute more evenly during learning.

**Benefits:**

- Faster convergence
- More stable gradients
- Better optimization
- Improved numerical stability

# Types of Normalization

There are several normalization techniques.

## 1. Min-Max Normalization

Transforms values into a fixed range, usually:

$$
[0,1]
$$

Formula:

$$
x'=
\frac{x-x_{min}}
{x_{max}-x_{min}}
$$

Example:

Original:

$$
[10,20,30]
$$

Normalized:

$$
[0,0.5,1]
$$

# 2. Z-Score Standardization

Centers data around zero and scales it to unit variance.

Formula:

$$
x'
=
\frac{x-\mu}
{\sigma}
$$

where:

$$
\mu = \text{mean}
$$

$$
\sigma = \text{standard deviation}
$$

Example:

Original:

$$
[2,4,6,8]
$$

Mean:

$$
5
$$

Standard deviation:

$$
2.236
$$

Normalized:

$$
[-1.34,-0.45,0.45,1.34]
$$

# Normalization in Deep Learning

In deep neural networks, activations can become very large or very small.

Example:

Layer 1:

$$
[2,4,6]
$$

Layer 2:

$$
[200,400,600]
$$

Layer 3:

$$
[0.002,0.004,0.006]
$$

This causes:

- Exploding activations
- Vanishing activations
- Unstable gradients

Normalization solves this problem.

# Types of Normalization in Deep Learning

## Batch Normalization

Introduced in 2015.

Normalizes across examples in a mini-batch.

Formula:

$$
\hat{x}
=
\frac{x-\mu_B}
{\sqrt{\sigma_B^2+\epsilon}}
$$

where:

$$
\mu_B
=
\text{batch mean}
$$

$$
\sigma_B^2
=
\text{batch variance}
$$

Used primarily in:

- CNNs
- Computer Vision

# Layer Normalization

Instead of normalizing across a batch, LayerNorm normalizes across features of a single token.

Input:

$$
x=[2,4,6,8]
$$

Mean:

$$
\mu=5
$$

Variance:

$$
\sigma^2=5
$$

LayerNorm:

$$
LayerNorm(x)
=
\frac{x-\mu}
{\sqrt{\sigma^2+\epsilon}}
\gamma+\beta
$$

Used in:

- Transformers
- BERT
- GPT

# Instance Normalization

Normalization is performed separately for each sample and channel.

Commonly used in:

- Style Transfer
- Image Generation

# Group Normalization

Channels are divided into groups.

Normalization is applied within each group.

Used when:

- Batch sizes are very small
- BatchNorm becomes unstable

# RMS Normalization (Modern LLMs)

Used by:

- Llama
- Mistral
- Qwen
- DeepSeek

Unlike LayerNorm, RMSNorm does not subtract the mean.

Formula:

$$
RMS(x)
=
\sqrt{
\frac{1}{n}
\sum_{i=1}^{n}
x_i^2
}
$$

Normalization:

$$
RMSNorm(x)
=
\frac{x}{RMS(x)}
\gamma
$$

Advantages:

- Simpler
- Faster
- Lower computation
- Works well for very large models
